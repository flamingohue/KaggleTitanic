```python
from crr_common import crr_utils as cu
import pyspark.sql.functions as F
import pandas as pd
import numpy as np
import math
pd.set_option('display.max_rows', None)
import matplotlib.pyplot as plt
```

```python
#getting the datasets ready
org = pd.read_csv('output_data/org_score_table_20210430.csv')
ind=pd.read_csv('output_data/ind_score_table_20210430.csv')
org_orig = pd.read_csv('org_predicted_results.csv')
ind_orig = pd.read_csv('ind_predicted_result_noname.csv')
```

```python
org_orig.dtypes
```

```python
def Balance_Score(row_bal, row_lob, party_type):
    if (party_type=='ind'):
        if row_lob=='JFL_PIC':
            val = (row_bal -ind_pic['log_AvgBal'].mean())/ind_pic['log_AvgBal'].std()
        else: #trust 
            val = (row_bal -ind_iam['log_AvgBal'].mean())/ind_iam['log_AvgBal'].std()
    else:
        if row_lob=='JFL_PIC':
            val = (row_bal -org_pic['log_AvgBal'].mean())/org_pic['log_AvgBal'].std()
        else: 
            val = (row_bal -org_iam['log_AvgBal'].mean())/org_iam['log_AvgBal'].std()

    return val
```

```python
org['log_AvgBal']=org['Average_Balance'].map(lambda x: np.log(x) if x>=1 else 0)
ind['log_AvgBal']=ind['Average_Balance'].map(lambda x: np.log(x) if x>=1 else 0)
```

```python
ind_pic=ind[ind['aml_party_segment']=='JFL_PIC']
ind_iam=ind[ind['aml_party_segment']!='JFL_PIC']
org_pic=org[org['aml_party_segment']=='JFL_PIC']
org_iam=org[org['aml_party_segment']!='JFL_PIC']
```

```python
print(ind_pic['log_AvgBal'].mean())
print(ind_pic['log_AvgBal'].std())
print(ind_iam['log_AvgBal'].mean())
print(ind_iam['log_AvgBal'].std())
print(org_pic['log_AvgBal'].mean())
print(org_pic['log_AvgBal'].std())
print(org_iam['log_AvgBal'].mean())
print(org_iam['log_AvgBal'].std())
```

```python
org['Balance_score']=org.apply(lambda row: Balance_Score(row['log_AvgBal'], row['aml_party_segment'], 'org'), axis=1)
ind['Balance_score']=ind.apply(lambda row: Balance_Score(row['log_AvgBal'], row['aml_party_segment'], 'ind'), axis=1)
```

```python
org['Balance_Grp']=org['Balance_score'].map(lambda x:  (math.floor(x)))
org['Balance_Grp'].loc[org['Balance_Grp']<=-2]=-2
org['Balance_Grp'].loc[org['Average_Balance']<=1]=-999
ind['Balance_Grp']=ind['Balance_score'].map(lambda x:  (math.floor(x)))
ind['Balance_Grp'].loc[ind['Balance_Grp']<=-2]=-2
ind['Balance_Grp'].loc[ind['Average_Balance']<=1]=-999
```

```python
bal_dic={
    -999:"<=1 dollar",
    -2:'-2 std',
    -1:'-1 std',
    0: '0 std',
    1:'1 std',
    2:'2 std',
    3:'3 std'
}
```

### Org

```python
org['Balance_Grp']=org['Balance_Grp'].map(bal_dic)
org_orig['Balance_Grp']=org_orig['Balance_Grp'].map(bal_dic)
column_names = [ "JFL", "CB Wealth"]
org_bal = pd.DataFrame(columns = column_names)
org_bal['CB Wealth']=org_orig['Balance_Grp'].value_counts().sort_index()
org_bal['JFL']=org['Balance_Grp'].value_counts().sort_index()
#org_bal['JFL']=(org_bal['JFL']/org.shape[0])
#org_bal['CB Wealth']=(org_bal['CB Wealth']/org_orig.shape[0])
org_bal=org_bal.transpose()
```
